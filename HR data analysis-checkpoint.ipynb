```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv("HR Data.csv")
```

```python
# Initial data exploration
print(data.head())
print(data.info())
print(data.describe())
```

```python
# Data Cleaning
# Removing unnecessary columns (assuming some columns might be unnecessary, adjust as needed)
data = data.drop(columns=['Unnamed: 0'], errors='ignore')
```

```python
# Renaming columns for clarity
data.columns = [col.strip().lower().replace(' ', '_') for col in data.columns]
```

```python
# Dropping duplicate entries
data = data.drop_duplicates()
```

```python
# Handling missing values
for column in data.columns:
    if data[column].dtype == 'object':
        # Fill categorical columns with mode
        data[column] = data[column].fillna(data[column].mode()[0])
    elif data[column].dtype in ['int64', 'float64']:
        # Fill numerical columns with median
        data[column] = data[column].fillna(data[column].median())
```

```python
# Additional sanitization if needed (e.g., fixing data types)
# Example: Converting date columns to datetime format
if 'hire_date' in data.columns:
    data.loc[:, 'hire_date'] = pd.to_datetime(data['hire_date'], errors='coerce')
```

```python
# 2. Standardize categorical data
for column in data.select_dtypes(include='object').columns:
    data[column] = data[column].str.strip().str.lower().str.replace(' ', '_')
```

```python
# 3. Outlier treatment for numerical columns
for column in data.select_dtypes(include=['int64', 'float64']).columns:
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    data[column] = np.where(data[column] < lower_bound, lower_bound, data[column])
    data[column] = np.where(data[column] > upper_bound, upper_bound, data[column])
```

## Analysis

```python
# 1. Employee demographics
print('Employee Demographics:')
print(data['department'].value_counts())
print(data['gender'].value_counts())
print(data['department'].value_counts())
```

```python
# 2. Turnover analysis
if 'attrition' in data.columns:
    turnover_rate = data['attrition'].value_counts(normalize=True) * 100
    print('Turnover Rate:')
    print(turnover_rate)
```

```python
# 3. Correlation analysis (using only numerical columns)
numerical_data = data.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numerical_data.corr()
```

```python
# Extract top correlations (e.g., top 10)
corr_pairs = correlation_matrix.unstack().sort_values(key=abs, ascending=False)
top_corr = corr_pairs[corr_pairs.index.get_level_values(0) != corr_pairs.index.get_level_values(1)].drop_duplicates().head(10)
print('Top Correlation Pairs:')
print(top_corr)
```

```python
# Visualize only the top correlations
top_corr_matrix = correlation_matrix.loc[top_corr.index.get_level_values(0), top_corr.index.get_level_values(1)]
plt.figure(figsize=(10, 6))
sns.heatmap(top_corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Top Correlation Heatmap', fontsize=20)
plt.tight_layout()
plt.show()
```

```python
# Visualization
plt.figure(figsize=(14, 8))
sns.countplot(x='department', hue='attrition', data=data, palette='Set2')
plt.title('Attrition by Department', fontsize=20)
plt.xlabel('Department', fontsize=16)
plt.ylabel('Count', fontsize=16)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.countplot(x='gender', hue='attrition', data=data, palette='Set1')
plt.title('Attrition by Gender', fontsize=20)
plt.xlabel('Gender', fontsize=16)
plt.ylabel('Count', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Example of alternative analysis if 'monthly_income' does not exist
if 'age' in data.columns and 'attrition' in data.columns:
    plt.figure(figsize=(16, 9))
    sns.boxplot(x='department', y='age', hue='attrition', data=data, palette='viridis')
    plt.title('Age vs Department by Attrition', fontsize=20)
    plt.xlabel('Department', fontsize=16)
    plt.ylabel('Age', fontsize=16)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
else:
    print('Column "age" not found for analysis.')
```
